```python
import matplotlib.pyplot as plt
```

```python
model_names = ['EfficientNetB6', 'ResNet50', 'VGG16']
mean_validation_accuracies = [0.308, 0.296, 0.504]
training_times = [35, 5, 11]  # in minutes

training_metrics = {
    'EfficientNetB6': {'loss': [0.4171, 0.2251], 'accuracy': [0.8547, 0.9231]},
    'ResNet50': {'loss': [0.6130, 0.3439], 'accuracy': [0.7996, 0.8793]},
    'VGG16': {'loss': [1.2664, 0.9821], 'accuracy': [0.4197, 0.5672]}
}
```

```python
fig, ax1 = plt.subplots(figsize=(12, 10))
color = 'tab:blue'
ax1.set_xlabel('Model')
ax1.set_ylabel('Mean Validation Accuracy', color=color)
ax1.bar(model_names, mean_validation_accuracies, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Training Time (minutes)', color=color)
ax2.plot(model_names, training_times, color=color, marker='o', linestyle='--')
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:green'
ax3.spines['right'].set_position(('outward', 60))
ax3.set_ylabel('Training Loss & Accuracy', color=color)
for model_name in model_names:
    ax3.plot([model_name]*2, training_metrics[model_name]['loss'], color=color, marker='x', linestyle='-', label=f'{model_name} Loss')
    ax3.plot([model_name]*2, training_metrics[model_name]['accuracy'], color=color, marker='s', linestyle='-', label=f'{model_name} Accuracy')
ax3.tick_params(axis='y', labelcolor=color)

lines, labels = ax3.get_legend_handles_labels()
ax3.legend(lines, labels, loc='lower right')
plt.title('Performance of Different Models')
plt.tight_layout()
plt.grid(alpha=0.4)
plt.show()
```
